```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as mpl

x = range (-100000, 100000)
y1=[]
y2=[]
y3=[]
y4=[]
y5=[]
y6=[]

sample = pd.DataFrame()

def make_data(num2, num3, num1,y):
    for i in range(-100000, 100000):
        x = i
        n = 0
        while (1):
            if x == 1:
                y.append(n)
                break
            elif n >= 1000:
                y.append(0)
                break
                
            elif x/2 == int(x/2) and x != 0 :
                x /= num2
                n += 1
            
                
            
            else:
                x = x*num3+num1
                n += 1
def plot():
    fig = mpl.figure()
    mpl.xlabel("x")
    mpl.ylabel("y")
    mpl.grid(True)
    mpl.scatter(x,y1,s=0.01,c="Red", label="(+2,+3,1)") 
    mpl.legend()
    fig.savefig("y1.png")
    mpl.show()
    fig = mpl.figure()
    mpl.scatter(x,y2,s=0.01, c="Blue", label="(2,3,-1)") 
    mpl.legend()
    fig.savefig("y2.png")
    mpl.show()
    fig = mpl.figure()
    mpl.scatter(x,y3,s=0.01, c="Green", label="(-2,3,1)")
    mpl.legend()
    fig.savefig("y3.png")
    mpl.show()
    fig = mpl.figure()
    mpl.scatter(x,y4,s=0.01, c="Yellow", label="(-2,3,-1)")
    mpl.legend()
    fig.savefig("y4.png")
    mpl.show()
    fig = mpl.figure()
    mpl.scatter(x,y5,s=0.01, c="Black", label="(-2,-3,1)")
    mpl.legend()
    fig.savefig("y5.png")
    mpl.show()
    fig = mpl.figure()
    mpl.scatter(x,y6,s=0.01, c="Brown", label="(-2,-3,-1)")
    mpl.legend()
    fig.savefig("y6.png")
    mpl.show()

def make_table():
    data = np.array([x,y1,y2,y3,y4,y5,y6])
    df = pd.DataFrame(data, index=["","(+2,+3,1)", "(2,3,-1)", "(-2,3,1)", "(-2,3,-1)","(-2,-3,1)","(-2,-3,-1)"])
    df = df.T
    sample = df[100000:199999]
    print(sample.sample(20))

make_data(+2,3,1 ,y1)
make_data(2,3,-1,y2)
make_data(-2,3,1 ,y3)
make_data(-2, 3,-1, y4)
make_data(-2, -3,1, y5)
make_data(-2, -3,-1, y6)
plot()
make_table()
```
